### Handle outliers and wrong values

Firstly we plot the different features

The *SSM1km-ssm* and *FCOVER300-RT0-FCOVER* present constant values, this could be a problem of the original dataset but we're going to leave until further information.

#### Albedo

For the albedo features (*ALBH* ans *ALDH*) the maximum value should be 10000 and some specif values are used:

- 65535 for invalid values
- 65534 for values with out of range value inferior to the minimum physical value
- 65533 for values with out of range value superior to the max physical value

To align the values we will firstly set 65534 to 0, 65533 and every other value greater than 10000 to 10000

```python
albedo_columns = ['ALBH-AL-BH-NI', 'ALBH-AL-BH-VI', 'ALBH-AL-BH-BB',
       'ALDH-AL-DH-BB', 'ALDH-AL-DH-VI', 'ALDH-AL-DH-NI']
df_land_albedo = df_land[albedo_columns].copy()

df_land_albedo[df_land_albedo == 65534] = 0
df_land_albedo[df_land_albedo > 10000] = 10000

df_land_albedo.describe()
```

To detect the outliers we will use the quartile and interquartile range and define a value as outliers if

$v < Q1 - 1.5 * IQR or v > Q3 + 1.5 * IQR$

```python
Q1 = df_land_albedo.quantile(0.25)
Q3 = df_land_albedo.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

Count of outliers:

```python
mask = df_land_albedo[(df_land_albedo < (Q1-1.5*IQR)) | (df_land_albedo > (Q3+1.5*IQR))].isna()

df_land_albedo[(df_land_albedo < (Q1-1.5*IQR)) | (df_land_albedo > (Q3+1.5*IQR))]
```

Now that we know what are the outliers, we can correct them by taking the average of the 4 closest coordinates

```python
#### ADJUST TO NOT CONSIDER THE OUTLIERS

def get_avg_value(dataset, mask, lon, lat, precision):
    up = dataset.loc[lon, lat+precision]
    down = dataset.loc[lon, lat-precision]
    right = dataset.loc[lon+precision, lat]
    left = dataset.loc[lon-precision, lat]
    print(up, down, right, left)

    
```

## Merge of Climate and Land dataset

```python
# The combined index is composed as (latitude*100)(longitude*100)
# x*yyyyy 
# where x* is a variable length number representing latitude*100
# and yyyyy is a 5 digits longitude*100 with zero padding

combine_index = (lambda x: int(str(int(x[0]*100))+str(int(x[1]*100)).zfill(5)))
df1 = df_cds.copy()
df2 = df_land.copy()

df1.index = df1.index.map(combine_index)
df1.index.name = "latitude-longitude"
df2.index = df2.index.map(combine_index)
df2.index.name = "latitude-longitude"

df = pd.merge_asof(df2.sort_values(['latitude-longitude']), df1, on='latitude-longitude', allow_exact_matches=False)
```
